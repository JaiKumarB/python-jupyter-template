### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

def plot_subplots(x_line, y_line, categories, values, x_scatter, y_scatter, labels_pie, sizes_pie):
    """Plot a 2x2 subplot layout with error handling for dataset lengths."""
    try:
        
        if len(x_line) != len(y_line):
            raise ValueError("Line plot: Length of x and y must match.")
        
       
        if len(categories) != len(values):
            raise ValueError("Bar plot: Number of categories must match number of values.")
        
       
        if len(x_scatter) != len(y_scatter):
            raise ValueError("Scatter plot: Length of x and y must match.")
        
        
        if len(labels_pie) != len(sizes_pie):
            raise ValueError("Pie chart: Number of labels must match number of sizes.")
        if sum(sizes_pie) != 100:
            raise ValueError("Pie chart: The sum of sizes must equal 100.")

        # Creating subplots
        fig, axs = plt.subplots(2, 2, figsize=(10, 10))

        # Line plot
        axs[0, 0].plot(x_line, y_line, color='blue')
        axs[0, 0].set_title('Line Plot')
        axs[0, 0].set_xlabel('x')
        axs[0, 0].set_ylabel('y')
        axs[0, 0].grid(True)

        # Bar plot
        axs[0, 1].bar(categories, values, color=['#FF5733', '#33FF57', '#3357FF', '#F333FF'])
        axs[0, 1].set_title('Bar Plot')
        axs[0, 1].set_xlabel('Categories')
        axs[0, 1].set_ylabel('Values')

        # Scatter plot
        axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
        axs[1, 0].set_title('Scatter Plot')
        axs[1, 0].set_xlabel('x')
        axs[1, 0].set_ylabel('y')

        # Pie chart
        axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140, colors=['#FF5733', '#33FF57', '#3357FF', '#F333FF'])
        axs[1, 1].set_title('Pie Chart')

        # Display
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error: {e}")



x_line = list(range(0, 11))
y_line = [x**2 for x in x_line]

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

plot_subplots(x_line, y_line, categories, values, x_scatter, y_scatter, labels_pie, sizes_pie)
```
