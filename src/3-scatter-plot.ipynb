### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

def plot_scatter(x, y):
 
    try:
       
        if not isinstance(x, list) or not isinstance(y, list):
            raise TypeError("Both x and y should be lists.")
        if len(x) != len(y):
            raise ValueError("The lengths of x and y should match.")
        if not all(isinstance(i, (int, float)) for i in x + y):
            raise ValueError("All elements in x and y must be numeric.")
      
        plt.scatter(x, y, color='blue', label='Data Points')
        for i, (x_val, y_val) in enumerate(zip(x, y)):
            plt.annotate(f'({x_val}, {y_val})', (x_val, y_val), textcoords="offset points", xytext=(5,5), ha='center')
        
       
        plt.xlabel('x values')
        plt.ylabel('y values')
        plt.title('Scatter Plot with Annotations')
        plt.grid(True)
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"Error: {e}")


x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


plot_scatter(x, y)
```
