### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

def plot_pie(labels, sizes):
    """Plot a pie chart with error handling for input data."""
    try:
       
        if not isinstance(labels, list) or not isinstance(sizes, list):
            raise TypeError("Both labels and sizes should be lists.")
        if len(labels) != len(sizes):
            raise ValueError("The number of labels and sizes must match.")
        if not all(isinstance(s, (int, float)) for s in sizes):
            raise ValueError("All elements in sizes must be numeric.")
        if sum(sizes) != 100:
            raise ValueError("The sum of sizes must equal 100.")

        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#FF5733', '#33FF57', '#3357FF', '#F333FF'])
        plt.title('Programming Language Usage')
        plt.show()

    except Exception as e:
        print(f"Error: {e}")


labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]


plot_pie(labels, sizes)
```
